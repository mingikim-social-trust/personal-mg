```python
conn.rollback()

qry = """
SELECT
    b.name,
    COUNT(*) AS total_count,
    SUM(
        CASE 
            WHEN successAcc."bankAccountId" IS NULL
                 AND (bpr.status = 'timeout' OR bpr."processedAt" IS NULL)
            THEN 1 
            ELSE 0 
        END
    ) AS timeout_count,
    SUM(
        CASE 
            WHEN bpr.status = 'completed' THEN 1 
            ELSE 0 
        END
    ) AS success_count,
    COUNT(DISTINCT ba."userId") AS user_count
FROM bs_parse_request bpr
LEFT JOIN bank_account ba 
       ON bpr."bankAccountId" = ba."id"
LEFT JOIN bank b 
       ON ba."bankId" = b."id"

LEFT JOIN (
    SELECT DISTINCT "bankAccountId"
    FROM bs_parse_request
    WHERE status = 'completed'
) AS successAcc
ON successAcc."bankAccountId" = bpr."bankAccountId"

WHERE bpr."createdAt" >= '2025-01-17'
GROUP BY b.name;

        """
cursor.execute(qry)
result = cursor.fetchall()

# DataFrame 생성
df = pd.DataFrame(result, columns=[desc[0] for desc in cursor.description])
df.to_csv('./timeout_count_recent_이미제출과중복제거.csv', index=False, encoding='utf-8-sig')

# 결과 확인
print(df.head())
```
